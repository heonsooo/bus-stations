```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
import pandas as pd
import seaborn as sns
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 윈도우의 한글 폰트 설정
plt.rc('font', family = 'Malgun Gothic')

# 시각화 그래프가 노트북 안에 보이게 하기
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

```python
bus_2019 = os.listdir('./data/2019')
bus_2020 = os.listdir('./data/2020_1')
```

# 2019년도 버스 data

```python
df = pd.read_csv('./data/2019/BUS_STATION_BOARDING_MONTH_201901.csv',encoding='euc-kr')
```

```python
for bus_month in bus_2019[1:]:
    df = df.append(pd.read_csv('./data/2019/'+bus_month,encoding='euc-kr'))
```

```python
df[df['노선번호'].str.contains('271')]
```

```python
df_271 = df[df['노선번호'].str.contains('271')]
```

# 2020년도

```python
df_2020 = pd.read_csv('./data/2020_1/'+bus_2020[0],encoding='euc-kr')
```

```python
df_2020
```

```python
for bus_month in bus_2020[1:]:
    df_2020 = df_2020.append(pd.read_csv('./data/2020_1/'+bus_month,encoding='euc-kr'))
```

```python
df_2020.head()
```

```python
df_2020['노선번호'] = df_2020['노선번호'].astype('str') 
```

```python
df_271 = df_271.append(df_2020[df_2020['노선번호'].str.contains('271')])
```

```python
df_271.노선번호.unique()
```

```python
df_271 =df_271.reset_index(drop= True)
```

```python
df_271= df_271.drop(columns=['노선명','버스정류장ARS번호','등록일자','노선ID','표준버스정류장ID','역ID'])
```

```python
df_271['총승객수'] = df_271['승차총승객수'] + df_271['하차총승객수']
```

```python
df_271['사용일자'].nunique()
```

```python
df_271
```

```python
month_ = []
for i in list(df_271['사용일자']):
    month_.append(int(str(i)[:6]))
```

```python
df_271['Month'] = month_
```

```python
df_271.Month= pd.to_datetime(df_271.Month, format='%Y%m')
```

```python
df_271_passengers_by_month = df_271.groupby(
    by=['Month', '노선번호']).sum('총승객수').drop(columns=['승차총승객수', '하차총승객수','사용일자'])
```

```python
df_271_passengers_by_month.head()
```

```python
df_271_passengers_by_month.reset_index(inplace=True)
df_271_passengers_by_month.head()
```

```python
df_271_passengers_by_month['노선번호'] = df_271_passengers_by_month['노선번호'].astype('str')
```

```python
df_271_passengers_by_month.head()
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize=(16,9))
sns.lineplot(data=df_271_passengers_by_month, x='Month',y='총승객수',hue='노선번호',)
```

# 전체 271

```python
bus_271A = df_271_passengers_by_month[df_271_passengers_by_month['노선번호'].str.contains('271A')]
bus_271B = df_271_passengers_by_month[df_271_passengers_by_month['노선번호'].str.contains('271B')]
```

```python
# 분할 후 271 승객 수 
bus_271 = bus_271A.append(bus_271B).groupby('Month').sum()
bus_271.reset_index(inplace=True)
bus_271['노선번호'] = [271]*len(bus_271)
```

```python
bus_271 = bus_271.append(df_271_passengers_by_month[df_271_passengers_by_month['노선번호']=='271'])
bus_271.reset_index(drop=True)
bus_271.tail(5)
```

```python
bus_271.Month= pd.to_datetime(bus_271.Month, format='%Y%m')
bus_271.head()
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize=(16,9))
sns.lineplot(data=bus_271,x='Month',y='총승객수')
```
